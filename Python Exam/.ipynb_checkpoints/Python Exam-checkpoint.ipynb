```python
class InvalidGradeError(Exception):
    """Custom exception for invalid grade values."""
    pass


def read_student_records(filename):
    try:
        with open(filename, "r") as file:
            grades = []
            for line in file:
                name, grade = line.strip().split(",")
                grade = grade.strip()

                
                if not grade.isdigit():
                    raise InvalidGradeError(f"Invalid grade for {name}: {grade}")

                grades.append(int(grade))

            
            if not grades:
                print("No valid grades found in the file.")
                return

            
            average = sum(grades) / len(grades)
            print(f"Average grade of students: {average:.2f}")

    except FileNotFoundError:
        print("Error: The file was not found. Please check the filename.")
    except InvalidGradeError as e:
        print(e)
    except Exception as e:
        print("An unexpected error occurred:", e)



filename = "students_record.txt"
read_student_records(filename)
```

```python
def perform_operations(numbers):
    total = sum(numbers)
    product = 1
    for n in numbers:
        product *= n
    average = total / len(numbers)
    return total, product, average

numbers = [85, 90, 75, 82, 95, 78]
total, product, average = perform_operations(numbers)
print("Total:", total)
print("Product:", product)
print("Average:", round(average, 2))
```

Question no 2

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print(f"{self.make} {self.model}'s engine has started.")

    def display_info(self):
        print(f"Make: {self.make}, Model: {self.model}, Year: {self.year}")


class Car(Vehicle):
    def __init__(self, make, model, year, doors):
        super().__init__(make, model, year)
        self.doors = doors

    def display_info(self):
        super().display_info()
        print(f"Number of Doors: {self.doors}")

    def accelerate(self):
        print(f"{self.make} {self.model} is accelerating!")


class Truck(Vehicle):
    def __init__(self, make, model, year, cargo_capacity):
        super().__init__(make, model, year)
        self.cargo_capacity = cargo_capacity

    def display_info(self):
        super().display_info()
        print(f"Cargo Capacity: {self.cargo_capacity} tons")

    def load_cargo(self):
        print(f"{self.make} {self.model} is loading {self.cargo_capacity} tons of cargo.")


class Motorcycle(Vehicle):
    def __init__(self, make, model, year, drive_type):
        super().__init__(make, model, year)
        self.drive_type = drive_type

    def display_info(self):
        super().display_info()
        print(f"Drive Type: {self.drive_type}")

    def wheelie(self):
        print(f"{self.make} {self.model} is doing a wheelie!")


car1 = Car("Toyota", "Corolla", 2022, 4)
truck1 = Truck("Volvo", "FH16", 2021, 25)
bike1 = Motorcycle("Yamaha", "R1", 2023, "Chain Drive")

print("\n--- Car Details ---")
car1.display_info()
car1.start_engine()
car1.accelerate()

print("\n--- Truck Details ---")
truck1.display_info()
truck1.start_engine()
truck1.load_cargo()

print("\n--- Motorcycle Details ---")
bike1.display_info()
bike1.start_engine()
bike1.wheelie()
```

```python
def perform_operations(numbers):
    total = 0
    product = 1
    
    for num in numbers:
        total += num
        product *= num
    
    average = total / len(numbers)
    return total, product, average

numbers = [1, 2, 3, 4, 5]
total, product, average = perform_operations(numbers)

print("Total:",total)
print("Product:",product)
print("Average:",average)
```
