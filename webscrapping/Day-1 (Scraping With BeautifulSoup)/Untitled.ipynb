```python
data_dict = {}

for i, j in enumerate(job_divs):
    title = j.find(class_='title').text
    sub_title = j.find(class_='subtitle').text
    location = j.find(class_='location').text.strip()
    date = j.find(class_='is-small').text.strip()

    data_dict[i] = {
        'title': title,
        'subtitle': sub_title,
        'location': location,
        'date': date
    }
```
