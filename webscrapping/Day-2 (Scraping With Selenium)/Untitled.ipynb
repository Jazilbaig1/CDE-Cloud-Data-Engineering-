```python
divs_data = []
start = 0

while start <= 3:
    print(f'Scrapping Page: {start}')
    divs = driver.find_elements(By.CLASS_NAME, 'quote')
    for i, div in enumerate(divs):
        # print(f'Div Index: {i}')
        title = div.find_element(By.CLASS_NAME, 'text')
        author = div.find_element(By.CLASS_NAME, 'author')
        tags = div.find_element(By.CLASS_NAME, 'tags')
        divs_data.append((title.text, author.text, tags.text))
    next_btn = driver.find_element(By.XPATH, '/html/body/div/nav/ul/li/a')
    next_btn.click
    start += 1
```

```python
import time
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
```

```python
position = 'data engineer'
location = 'islamabad'

url = f'https://pk.indeed.com/jobs?q={position.replace(' ', '+')}&l={location.replace(' ', '+')}'
url
```

```python
driver = uc.Chrome()
```

```python
driver.get(url)
```

```python
cards = driver.find_elements(By.CLASS_NAME, 'cardOutline')
```

```python
job_links = []

for card in cards[:4]:
    link = card.find_element(By.CSS_SELECTOR, '.jcs-JobTitle').get_attribute('href')
    job_links.append(link)
    print('Link', link)
```

```python
data = []

for i, link in enumerate(job_links):
    print('Scrapping: ', i)
    driver.get(link)
    time.sleep(5)

    job_title = driver.find_element(By.CLASS_NAME, 'jobsearch-JobInfoHeader-title')
    company = driver.find_element(By.XPATH, '//*[@id="viewJobSSRRoot"]/div/div[2]/div[3]/div/div/div[1]/div[2]/div[1]/div[2]/div/div/div/div[1]/div/span')
    location = driver.find_element(By.CLASS_NAME, 'css-1wiq240')
    desc = driver.find_element(By.ID, 'jobDescriptionText')
    data.append([job_title.text, company.text, location.text, desc.text])
print('data scrapping completed')
```
